```python
duration_mean = total_songs["duration_ms"].mean()
acustic_mean = total_songs["acousticness"].mean()
dance_mean = total_songs["danceability"].mean()
energy_mean = total_songs["energy"].mean()
intru_mean = total_songs["instrumentalness"].mean()
key_mean = total_songs["key"].mean()
live_mean = total_songs["liveness"].mean()
loud_mean= total_songs["loudness"].mean()
mode_mean = total_songs["mode"].mean()
speech_mean = total_songs["speechiness"].mean()
tempo_mean = total_songs["tempo"].mean()
valance_mean = total_songs["valence"].mean()
```
